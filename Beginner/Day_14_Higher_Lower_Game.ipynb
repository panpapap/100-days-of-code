```python
# Compare the follower_count
def check_answer(answer, a_followers, b_followers):
  """Checks followers against user's guess 
  and returns True if they got it right.
  Or False if they got it wrong.""" 
  if a_followers > b_followers:
    return answer == "a"
  else:
    return answer == "b"
```

```python
#Function to choose a random data
def rand_topic(the_data):
  return random.choice(data)
```

```python
#The game
def game():
  #print the logo
  print(logo)
  should_continue = True
  score = 0 

  #choose the A and B
  chosen_a = rand_topic(data)
  chosen_b = rand_topic(data)

  while should_continue:
    #Make the second topic be the first one for every round
    chosen_a = chosen_b
    chosen_b = rand_topic(data)
    #If the 2 topics got overlapped, then we will choose a different topic B 
    while chosen_a == chosen_b:
      chosen_b = rand_topic(data)

    print(f"Compare A: {chosen_a['name']}, {chosen_a['description']}, from {chosen_a['country']}")
    print(vs)
    print(f"Against B: {chosen_b['name']}, {chosen_b['description']}, from {chosen_b['country']}" )

    #Ask for the player's answer
    answer = input("Who has more followers? Type 'A' or 'B'.").lower()
    a_follower_count = chosen_a["follower_count"]
    b_follower_count = chosen_b["follower_count"]
    is_correct = check_answer(answer, a_follower_count, b_follower_count)

  #check if the guess is correct or not
    if is_correct:
      score += 1
      print(f"You're right! Current score: {score}.")
    else:
      should_continue = False
      print(f"Sorry, that's wrong. Final score: {score}")

game()
```
